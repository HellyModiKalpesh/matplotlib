```python
# Explosion values:Distance of each slice from the center (explode effect)
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["c++","c","python","java"]
ex=[0.2,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex)
plt.show()
```

```python
# percentage value

#autopct="%0.1f%%": Format string for the percentages shown on each slice.
#%0.1f%% formats the number to one decimal place followed by a percentage sign.
#shadow=True it will create 3d effect
#radius:It will increase the radius of pie chart
#labeldistace:Distance of the labels from centre
#startangle:Starting angle for the pie chart
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["c++","c","python","java"]
ex=[0.2,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",shadow=True,radius=1.2,labeldistance=1.2,startangle=180)
plt.show()
```

```python
#Textprops:Dictionary specifying text properties such as font size
#counterclock:Draws slices in clockwise direction
# wedgeprops:Dictionary specifying wedge properties like border width and color
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["c++","c","python","java"]
ex=[0.2,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",radius=1.3,labeldistance=1.1,textprops={"fontsize":15},counterclock=False,wedgeprops={"linewidth":4,"edgecolor":"black"})
plt.show()
```

```python
# rotatelabels: rotate the labels
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["c++","c","python","java"]
ex=[0.2,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",radius=1.1,labeldistance=1.1,textprops={"fontsize":15},counterclock=False
        ,wedgeprops={"linewidth":4},rotatelabels=True)

plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
x1=[40,30,20,10]
y=["c++","c","python","java"]
c=["r","b","y","g"]

plt.pie(x,labels=y,radius=1.5)
plt.pie(x1,radius=1,colors=c)


plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
x1=[40,30,20,10]
y=["c++","c","python","java"]
c=["r","b","y","g"]

plt.pie(x,labels=y,radius=1.3)
plt.pie([1],colors="white")

plt.show()
```

```python
# different way to create ring pie chart
# plt.Circle(xy=(0,0), radius=1, facecolor="w":This creates a white circle object (cr) with its center at (0,0) and a radius of 1.
# plt.gca() stands for "get current axes". It returns the current Axes instance on the current figure.
# .add_artist(cr) adds the white circle we created earlier to these axes.
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
x1=[40,30,20,10]
y=["c++","c","python","java"]
c=["r","b","y","g"]
cr=plt.Circle(xy=(0,0),radius=1,facecolor="w")


plt.pie(x,labels=y,radius=1.3)
plt.gca().add_artist(cr)
plt.show()
```
