```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,3,2,4,5]
plt.step(x,y,marker="o",color="red",ms=10,mfc="g",label="step plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("step demonstration")

plt.grid()
plt.legend(loc=2)
plt.show()
```
