```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,3,2,5,4]
z=[5,6,2,3,1]
t=[2,5,6,7,1]
l=["y","z","t"]
plt.stackplot(x,y,z,t,labels=l)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,3,2,5,4]
z=[5,6,2,3,1]
t=[2,5,6,7,1]
l=["y","z","t"]
plt.stackplot(x,y,z,t,labels=l,colors=["r","m","y"])
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
#plt.grid()
plt.show()
```
