```python
x=["c","python","java","c++"]
y=[60,90,90,50]
c=["y",'r',"b",'g']
plt.xlabel("languages",fontsize=15)
plt.ylabel("popularity",fontsize=15)
plt.title("survey",fontsize=15)
plt.bar(x,y,width=0.4,color=c,align="center",edgecolor='black',linewidth=5)
plt.show()
```

```python
x=["c","python","java","c++"]
y=[60,90,90,50]
c=["y",'r',"b",'g']
plt.xlabel("languages",fontsize=15)
plt.ylabel("popularity",fontsize=15)
plt.title("survey",fontsize=15)
plt.bar(x,y,width=0.4,color=c,align="center",edgecolor='black',linewidth=5,alpha=0.5)
plt.show()
```

```python
x=["c","python","java","c++"]
y=[60,90,90,50]
c=["y",'r',"b",'g']
plt.xlabel("languages",fontsize=15)
plt.ylabel("popularity",fontsize=15)
plt.title("survey",fontsize=15)
plt.bar(x,y,width=0.4,color=c,edgecolor='black',label="languages popularity")
plt.legend()
plt.show()
```

```python
x=["c","python","java","c++"]
y=[60,90,90,50]
z=[40,50,60,70]
c=["y",'r',"b",'g']
plt.xlabel("languages",fontsize=15)
plt.ylabel("popularity",fontsize=15)
plt.title("survey",fontsize=15)
plt.bar(x,y,width=0.4,color="red",edgecolor='black',label="languages popularity")
plt.bar(x,z,width=0.4,color="yellow",edgecolor='black',label="languages popularity1")
plt.legend()
plt.show()
```

```python
import numpy as np

x=["c","python","java","c++"]
y=[60,90,90,50]
z=[40,50,60,70]
c=["y",'r',"b",'g']

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("languages",fontsize=15)
plt.ylabel("popularity",fontsize=15)
plt.title("survey",fontsize=15)

plt.bar(p,y,width,color="red",edgecolor='black',label="languages popularity")
plt.bar(p1,z,width,color="yellow",edgecolor='black',label="languages popularity1")

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
import numpy as np

x=["c","python","java","c++"]
y=[60,90,90,50]
z=[40,50,60,70]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("languages",fontsize=15)
plt.ylabel("popularity",fontsize=15)
plt.title("survey",fontsize=15)

plt.barh(x,y,width,color="red",edgecolor='black',label="languages popularity")
plt.barh(x,z,width,color="yellow",edgecolor='black',label="languages popularity1")

plt.legend()
plt.show()
```
