```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,3,4,3,7,1]
plt.stem(x,y)
plt.show()
```

```python
# lineformat
"-" - solid line
"--"-dashed line
"-.":dash-dot line
":"-Dotted line
# markerfmt="g*": Sets green circles ('g*') as markers for data points.
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,3,4,3,7,1]
plt.stem(x,y,linefmt=":",markerfmt="g*")
plt.show()
```

```python
# bottom=5: Specifies the baseline of the stem plot as 5.
# basefmt="black": Sets the color of the baseline to black.
# orientation='horizontal': Renders the stem plot horizontally.
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,3,4,3,7,1]
plt.stem(x,y,linefmt=":",markerfmt="go",bottom=5,basefmt="black",label="python",orientation='horizontal')
plt.legend()
plt.show()
```
