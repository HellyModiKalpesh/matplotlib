```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("data.csv")
ages=data['Age']
dev_salary=data['All_Devs']
py_salary=data['Python']
js_salry=data['JavaScript']

plt.plot(ages,py_salary,label="python")
plt.plot(ages,js_salry,label="javascript")
plt.plot(ages,dev_salary,label="All dev",linestyle='--',color="black")

plt.legend()
plt.title('Median salary by Age')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("data.csv")
ages=data['Age']
dev_salary=data['All_Devs']
py_salary=data['Python']
js_salry=data['JavaScript']

fig,(ax1,ax2) =plt.subplots(nrows=2,ncols=1,sharex=True)

ax1.plot(ages,dev_salary,label="All dev",linestyle='--',color="black")
ax2.plot(ages,py_salary,label="python")
ax2.plot(ages,js_salry,label="javascript")


ax1.legend()
ax1.set_title('Median salary by Age')
ax1.set_ylabel('Median salary')

ax2.legend()
ax2.set_xlabel("Ages")
ax2.set_ylabel('Median salary')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("data.csv")
ages=data['Age']
dev_salary=data['All_Devs']
py_salary=data['Python']
js_salry=data['JavaScript']

fig1,ax1 =plt.subplots()
fig2,ax2 =plt.subplots()

ax1.plot(ages,dev_salary,label="All dev",linestyle='--',color="black")
ax2.plot(ages,py_salary,label="python")
ax2.plot(ages,js_salry,label="javascript")


ax1.legend()
ax1.set_title('Median salary by Age')
ax1.set_xlabel("Ages")
ax1.set_ylabel('Median salary')

ax2.legend()
ax2.set_title('Median salary by Age')
ax2.set_xlabel("Ages")
ax2.set_ylabel('Median salary')

plt.tight_layout()
plt.show()
```
