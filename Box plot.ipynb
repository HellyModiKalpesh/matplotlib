```python
# vert=False : it will plot horizontal
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70]
plt.boxplot(x,vert=False,widths=0.4,labels=["python"])
plt.show()
```

```python
# patch_artist=True: Fills the boxes with colors.
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70]
plt.boxplot(x,labels=["python"],patch_artist=True,showmeans=True)
plt.show()
```

```python
# showmeans=True: Displays mean markers in the plot.
# sym="b+": Sets the outlier symbols to blue circles (b+).
# boxprops=dict(color="r"): Sets the color of the box outlines to red (r).
# capprops=dict(color="yellow"): Sets the color of the caps (end of whiskers) to yellow.
# whiskerprops=dict(color="m"): Sets the color of the whiskers to magenta (m).
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70,120,200]
plt.boxplot(x,labels=["python"],showmeans=True,whis=3.5,sym="b+",boxprops=dict(color="r"),capprops=dict(color="yellow"),whiskerprops=dict(color="m"))
plt.show()
```

```python
# two box plot
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70,120,200]
x1=[23,43,56,78,90,34,54,34,78]
plt.boxplot(x,labels=["python"],showmeans=True,whis=3.5,sym="b+",boxprops=dict(color="r"),capprops=dict(color="yellow"),whiskerprops=dict(color="m"))
plt.boxplot(x1,labels=["python"],showmeans=True,whis=3.5,sym="b+",boxprops=dict(color="r"),capprops=dict(color="yellow"),whiskerprops=dict(color="m"))

plt.show()
```

```python
# multiple box plot
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70,120,200]
x1=[23,43,56,78,90,34,54,34,78]
y=[x,x1]
plt.boxplot(y,labels=["python","java"],showmeans=True,sym="bo",boxprops=dict(color="r"),capprops=dict(color="yellow"),whiskerprops=dict(color="m"))

plt.show()
```
