```python
import matplotlib.pyplot as plt
```

```python
basic syntax
x=[]
y=[]
plt.scatter(x,y)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[3,4,8,6,3,8]
plt.scatter(x,y,color="red")
plt.title("scatter plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("Number",fontsize=20)

plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[3,4,8,6,3,8]
color = ["r", "b", "g", "m", "b", "orange"]
sizes=[100,200,500,300,400,560]
plt.scatter(x,y,c = color,s=sizes,alpha=1,marker='*',edgecolor="black",linewidth=1)
plt.title("scatter plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("Number",fontsize=20)

plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[3,4,8,6,3,8]

color = [10,49,30,29,56,34]
sizes=[100,200,500,300,400,560]
plt.scatter(x,y,c = color,s=sizes,cmap="BrBG")
plt.title("scatter plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("Number",fontsize=20)
t=plt.colorbar()
t.set_label("intensity")

plt.show()
```
